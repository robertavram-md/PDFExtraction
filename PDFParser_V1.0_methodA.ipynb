```python
# importing the required modules 
import PyPDF2
from PyPDF2 import PdfFileReader
from textacy import preprocessing
import ftfy
import textacy
from textacy import preprocessing
```

```python
pdf_text = ''
for i in range(0, pdfReader.numPages):
    pageObj = pdfReader.getPage(i)
    page = pageObj.extractText()
    pdf_text = pdf_text + ' ' + page 
```

```python
def read_pdf_extract_text(path='data/1.pdf'):
    pdfFileObj = open(path, 'rb') 

    # creating a pdf Reader object 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    pdf_text = ''
    print(pdfReader.numPages)
    for i in range(0, pdfReader.numPages):
        pageObj = pdfReader.getPage(i)
        page = pageObj.extractText()
        pdf_text = pdf_text + ' ' + page 
    
    text = ftfy.fix_text(pdf_text)
    data = ftfy.fix_encoding(text)
    data=textacy.preprocessing.remove.remove_punctuation(data, marks='\n')
    data=textacy.preprocessing.normalize.normalize_whitespace(data)
    data=textacy.preprocessing.normalize.normalize_unicode(data)
    data=textacy.preprocessing.remove.remove_accents(data)
    
    return pdf_text
```

```python
paths = ['data/1.pdf','data/2.pdf','data/3.pdf','data/4.pdf']
i = 0
for path in paths:
    i = i + 1
    print(path)
    d = read_pdf_extract_text(path)
    import pathlib
    output_path = 'data/' + str(i) + '.txt'
    pathlib.Path(output_path).write_text(d)
```
